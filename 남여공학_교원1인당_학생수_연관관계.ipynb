#남, 여, 공학 사고 발생 수 비교

```python
coeducation = pd.read_csv('/content/drive/MyDrive/데이터시각화/학교안전사고 경진대회/상위5위 주변학교 비교/남여공학비교.csv')
```

```python
# 데이터 필터링: 남녀공학과, 남자고등학교, 여자고등학교 데이터 추출
female_engineering = coeducation[coeducation['남여공학'] == '남여공학']
male_highschool = coeducation[coeducation['남여공학'] == '남자']
female_highschool = coeducation[coeducation['남여공학'] == '여자']

# 운동장 사고 수 비교를 위한 그래프 그리기 (예: 막대 그래프)
plt.figure(figsize=(8, 6))
school_types = ['남녀공학', '남자', '여자']
average_accidents = [
    female_engineering['운동장사고'].mean(),
    male_highschool['운동장사고'].mean(),
    female_highschool['운동장사고'].mean()
]

plt.bar(school_types, average_accidents)
plt.title('학교 유형별 평균 운동장 사고 수 비교')
plt.xlabel('학교 유형')
plt.ylabel('평균 운동장 사고 수')

# 그래프 표시
plt.show()
```
